```python

from nltk.tokenize import sent_tokenize

corpus = """
To tokenize text from a file in Python, you first need to read the file's content into a string
and then apply a tokenization method, such as using the nltk library's sent_tokenize function.
This is a standard initial step for many AI and NLP tasks.
"""

print(sent_tokenize(corpus))
```

```python
documents=sent_tokenize(corpus.strip())

print(documents)
print(type(documents))
```

```python
for sentence in documents:
    print(sentence)
```

```python
from nltk.tokenize import word_tokenize
```

```python
for sentence in documents:
    print(word_tokenize(sentence))
```

```python
word_tokenize(corpus)
```

```python
from nltk.tokenize import wordpunct_tokenize
```

```python
wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer
```

```python
tokenizer=TreebankWordTokenizer()
```

```python
tokenizer.tokenize(corpus)
```
