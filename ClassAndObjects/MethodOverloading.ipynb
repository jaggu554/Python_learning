```python
class Calculator:
    def addition(self,a,b=0,c=0):
        return a+b+c
    

cal=Calculator()
print(cal.addition(2,4))
print(cal.addition(2,3,4))
print(cal.addition(10))
```

```python
class Simple_Cal:
    def addition(self,*args):
        return sum(args)
    
add=Simple_Cal()
print(add.addition(2,3))
print(add.addition(2,3,4))
print(add.addition(1,2,3,4,5,6))
print(add.addition(10,20,3,4,5,6,7,9))
```
