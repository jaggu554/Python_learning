```python
class Person:
    def __init__(self,name,age,gender):
        self._name=name
        self._age=age
        self.gender=gender

class Employee(Person):
    def __init(self,name,age,gender):
        super().__init__(name,age,gender)


e=Employee("Jagadeeswar",101,"Male")

print(e._name)
print(e._age)
```
