```python
class Person:
    def __init__(self,name,age):
        self.age=age
        self.name=name

    def get_name(self):
        print(self.name)
    
    def set_name(self,name):
        self.name=name
    
    def get_age(self):
        print(self.age)

    def set_name(self,age):
        if self.age >0:
            self.age=age
        else:
            print("Negative age cannot entered...")
    def display(self):
        print(self.name,self.age)

p=Person("Jagadeeswar",25)
p.get_name()
p.set_name("Reddy")
p.get_name()
p.display()

```

```python
class Car:
    def __init__(self,brand,model):
        self.brand=brand
        self.model=model

    def display(self):
        print(f"{self.brand} {self.model}")

car=Car("Toyato",2021)

print(car.brand)
print(car.model)

car.display()
```

```python
class Car:
    def __init__(self,model,brand,engine):
        self.model=model
        self._brand=brand
        self._engine=engine
    

    def _show_details(self):
        print(f"Model :{self.model}, Brand :{self._brand}, Engine:{self._engine}")


class ElectricCar(Car):
    def __init__(self,model,brand,engine,battery):
        super().__init__(model,brand,engine)
        self.battery=battery

    def show_info(self):
        self._show_details()
        print(f"Battery Capacity :{self.battery}")

tesla =ElectricCar("Tesla","Model 5","Electric","batter 65")
tesla._show_details()
tesla.show_info()
```
