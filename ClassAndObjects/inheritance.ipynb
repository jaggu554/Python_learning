```python
class Car:
    def __init__(self,windows,doors,engine):
        self.windows=windows
        self.engine=engine
        self.doors=doors
    

car=Car(4,"petrol",5)
print(car.windows)
print(car.engine)
print(car.doors)
```

```python
class Tesla(Car):
    def __init__(self,windows,doors,engine,is_self_driving):
        super().__init__(windows,engine,doors)
        self.is_self_driving=is_self_driving

    def self_driving_car(self):
        return self.is_self_driving

t=Tesla(4,5,"Diesel",True)
print(t.self_driving_car())
print(t.doors)
print(t.engine)
print(t.windows)
```
