```python
class Dog:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    
    def bark(self):
        print(f"{self.name} woof!...")

dog=Dog("Jaggu",25)
print(dog.name)
dog.bark()
print(dog.age)
```

```python
class BankAccount:
    def __init__(self,owner,balance=0):
        self.owner=owner
        self.balance=balance
    
    def deposit(self,amount):
        self.balance+=amount
        print(f"The amount depoisted is :",self.balance)
    
    def withdraw(self,amount):
        if self.balance < amount:
            print("Insuffient Inbalance")
        else:
            self.balance-=amount
            print(f"Remaining Balance...",self.balance)

    def get_amount(self):
        return self.balance
    

bank=BankAccount("Jagadeeswar",5000)

```

```python
bank.withdraw(1000)
```

```python
bank.deposit(10000)
```

```python
bank.get_amount()
```
