```python
def print_key_value(**kwargs):
    for key,value in kwargs.items():
        print(f"{key} :{value}")

print_key_value(name="Jagadeeswar",age=30,city="Hyderabad",country="India")
```

```python
def print_details(*args,**kwargs):
    for i in args:
        print(i)

    for key,value in kwargs.items():
        print(f"{key} :{value}")

print_details(1,2,3,name="Jagadeeswar",age=30,city="Hyderabad")
```
