```python
nums = [1, 2, 3, 4, 5]
sum=0
for i in range(len(nums)):
    sum+=nums[i]
    nums[i]=sum

print(nums)
```

```python
n=int(input("Enter a number"))

rev=0;
temp=n
while n>0:
    rem=n%10
    rev=rev*10+rem
    
    n=n//10

print("original number : ",temp)
print("Reverse number : ",rev)

if rev==temp:
    print("palindrome")
else:
    print("Not palindrome")
```

```python
nums = [1, 2, 2, 3, 1, 4, 2]
dict={}
for i in nums:
    if i  not in dict:
        dict[i]=1
    else:
        dict[i]=dict.get(i)+1

print(dict)
```

```python
scores = [45, 78, 90, 32, 66, 88]
grades=[]
for i in scores:
    if i<60:
        continue
    if i>=85:
        grades.append('A')
    elif i>=70 and i<=84:
        grades.append('B')
    elif i>=60 and i<=69:
        grades.append('C')

print(grades)
```
